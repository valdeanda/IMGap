# Merging files

```python
merged1=pd.merge(df,reads[['Sample', 'Reads']], on='Sample')
merged1.head()
```

```python
merged2=pd.merge(merged1,df4[['Bin', 'Size']], on='Bin')
merged2.head()
```

```python
merged3=pd.merge(merged2,df3[['Original_Contig_Name', 'contigLen', 'Sample_Depth', 'Depth']],on='Original_Contig_Name')
```

```python
merged3.head()
```

# Rename to do a final merge

```python
df=merged3.rename(columns={"Sample": "Sampling_Site","Sample_Depth": "Sample",'Reads':'Total_Reads'})
```

```python
df.head()
```

# Merge

```python
df=pd.merge(df,reads[['Sample', 'Reads']],on='Sample')
```

```python
df.head()
```

# Compute abundance (scaffold lengh * depth/genome size)

```python
df['cov']=(df['contigLen'] *df['Depth'])/df['contigLen']
```

```python
df.head()
```

```python
df['Sum_cov'] = df['cov'].groupby(merged3['Bin']).transform('sum')
```

```python
df['NormalizedCov'] = df['Sum_cov']/df['Size']
```

```python
df.head()
```

```python
df['RelativeAbundance'] = df['NormalizedCov']/df['Reads']

df.head()
```

```python
df['RelativeAbundanceReadable'] = df['RelativeAbundance']*100000000
```

```python
df.head()
```

```python
df.columns
```

```python
finaldf=df[['Bin', 'Sample',
       'NormalizedCov', 'RelativeAbundance', 'RelativeAbundanceReadable']]
```

```python
finaldf.head()
```

```python
finaldf=finaldf.drop_duplicates()
```
